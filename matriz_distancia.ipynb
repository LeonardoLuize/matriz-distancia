```python
import spacy
import requests
import re
from bs4 import BeautifulSoup
import numpy as np
from traitlets.config.loader import Undefined

#criar uma variável para armazenar no modelo
nlp=spacy.load("en_core_web_sm")
print("ok")
```

```python
''' Limite para ir mais rápido'''
array_limit = 500
print_limit = 30

def print_array(array):
  for item_index in range(len(array)):
    if item_index < print_limit:
      print(array[item_index])

print("ok")
```

```python
# Leonardo Luize

#A página web (url) deve apontar para uma página web em inglês contendo, não menos que 1000 palavras.
#O texto desta página deverá ser transformado em um array de senteças.

site1 = requests.get("https://en.wikipedia.org/wiki/English_Wikipedia")
site2 = requests.get("https://en.wikipedia.org/wiki/Marilyn_Monroe")
site3 = requests.get("https://en.wikipedia.org/wiki/British_English")
site4 = requests.get("https://en.wikipedia.org/wiki/Elvis_Presley")
site5 = requests.get("https://en.wikipedia.org/wiki/Spanish_conquest_of_Pet%C3%A9n")

print("ok")
```

```python
# Leonardo Luize

''' 
Sua tarefa será gerar a matriz termo documento, dos documentos recuperados da internet e 
imprimir esta matriz na tela. Para tanto:

a) Considere que todas as listas de sentenças devem ser transformadas em listas de vetores, 
onde cada item será uma das palavras da sentença. 

b) Todos  os  vetores  devem  ser  unidos  em  um  corpus  único  formando  uma  lista  de  vetores, 
onde cada item será um lexema.  

c) Este único corpus será usado para gerar o vocabulário. 
'''

vocabulary = []
sentences = []

def get_site_texts(site):
    p_tags = BeautifulSoup(site.text, 'html.parser').find_all('p')

    for p in p_tags:
      p_text = nlp(p.get_text())
      for sentence in p_text.sents:
        sentences.append(sentence)

        for word in sentence:
          if word.text not in vocabulary and word.dep_ != "dep" and word.dep_ != "punct":
            vocabulary.append(word.text)

site_text1 = get_site_texts(site1)
site_text2 = get_site_texts(site2)
site_text3 = get_site_texts(site3)
site_text4 = get_site_texts(site4)
site_text5 = get_site_texts(site5)

if array_limit == 0:
  array_limit = len(sentences)

print("total de sentenças:", len(sentences))
print(vocabulary)
```

```python
''' 
d) O  resultado  esperado  será  uma  matriz  termo  documento  criada  a  partir  da  aplicação  da 
técnica bag of Words em todo o corpus.
'''

bag_of_words = [vocabulary]

def get_array_ocurrency(sentences, vocabulary):
  for sentence in sentences:
    word_frequency = []
    for word_vocabulary in vocabulary:
      count = 0

      for word in sentence:
        if word.text == word_vocabulary:
          count += 1

      word_frequency.append(count)

    bag_of_words.append(word_frequency)

get_array_ocurrency(sentences[:array_limit], vocabulary)

sentences.pop(1)
sentences.pop(1)
bag_of_words.pop(1)
bag_of_words.pop(1)

print_array(bag_of_words)
```

```python
''' 
Sua  tarefa  será  gerar  a  matriz  termo-documento  usando  TF-IDF  por  meio  da  aplicação  das 
fórmulas TF-IDF na matriz termo-documento criada com a utilização do algoritmo Bag of Words. Sobre 
o Corpus que recuperamos anteriormente. 
'''

DF = []

for sentence in sentences:
  i = 0
  for word2 in sentence:
    i += 1
  DF.append(i)

print(DF)
```

```python
tf = []
line_count = 0

for line in bag_of_words[1:array_limit]:
  tf.append([])

  for occurence in line:
    tf[line_count].append(occurence/DF[line_count])

  line_count += 1

idf = []

for word in range(len(bag_of_words[0])):
  count_occurence = 0

  for sentence in bag_of_words[1:array_limit]:
    if sentence[word] != 0:
      count_occurence += 1

  idf.append(np.log((len(sentences) + 1) / (count_occurence + 1)))

tf_idf = []
tf_idf_dictionaries = []

count_tf_sentences = 0

for tf_sentence in tf:
  tf_idf_dictionary = {}
  tf_idf.append([])

  for tf_value in tf_sentence:
    tf_idf_value = tf_value * idf[count_tf_sentences]
    tf_idf[count_tf_sentences].append(tf_idf_value)

  count_tf_sentences += 1

print_array(tf_idf)
```

```python
distance_matrix = []
until = 1

for tf_idf_sentence in tf_idf:
  distance = []
  a = tf_idf_sentence

  for second_tf_idf_sentence in tf_idf[0:until]:
    b = second_tf_idf_sentence

    linalg = np.linalg.norm(a) * np.linalg.norm(b)
    dot = np.dot(a,b)

    cos_sim = dot / linalg
    distance.append(cos_sim)

  distance_matrix.append(distance)
  until += 1

for i in range(len(distance_matrix)):
  if i < print_limit:
    print(distance_matrix[i])
```
